```python
import matplotlib.pyplot as plt
import numpy as np

#    
with open('analog_signal.txt', 'r') as file:
    analog_signal = [float(line) for line in file]

with open('digital_signal.txt', 'r') as file:
    digital_signal = [float(line) for line in file]

#  
plt.figure(figsize=(10, 6))

#   ()
plt.plot(analog_signal, label=' ', color='blue', linewidth=2)

#   ( )
plt.stem(digital_signal, label=' ', linefmt='red', markerfmt='ro', basefmt=' ')

#  
plt.title('   ')
plt.xlabel('')
plt.ylabel('')
plt.legend()
plt.grid(True)
plt.show()
```
